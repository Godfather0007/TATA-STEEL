```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
```

```python
df = pd.read_csv('TATASTEEL.NS.csv')
```

```python
df
```

```python

df = df[['Date', 'Close']]


df['Date'] = pd.to_datetime(df['Date'])


df.set_index('Date', inplace=True)


df.sort_index(ascending=True, inplace=True)


scaler = MinMaxScaler()
df['Close'] = scaler.fit_transform(df['Close'].values.reshape(-1, 1))


plt.figure(figsize=(12, 6))
plt.title('Normalized Stock Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Normalized Close Price')
plt.plot(df.index, df['Close'])
plt.show()
```

```python
train_size = int(len(df) * 0.8)
train_data = df.iloc[:train_size]
test_data = df.iloc[train_size:]
```

```python
def create_sequences(data, sequence_length):
    sequences = []
    target = []

    for i in range(len(data) - sequence_length):
        x = data[i:i+sequence_length]
        y = data[i:i+sequence_length]
        sequences.append(x)
        target.append(y)

    return np.array(sequences), np.array(target)

sequence_length = 10
X_train, y_train = create_sequences(train_data, sequence_length)
X_test, y_test = create_sequences(test_data, sequence_length)
```

```python
model = Sequential()

model.add(LSTM(50, return_sequences=True, input_shape=(X_train.shape[1], 1)))
model.add(LSTM(50, return_sequences=False))
model.add(Dense(25))
model.add(Dense(1))

model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
model.fit(X_train, y_train, batch_size=64, epochs=50)
```

```python
y_pred = model.predict(X_test)
```

```python
plt.figure(figsize=(12, 6))
plt.title('Stock Price Prediction')
plt.xlabel('Date')
plt.ylabel('Normalized Close Price')


plt.plot(test_data.index[sequence_length:], test_data['Close'][sequence_length:], label='Actual', color='blue')


plt.plot(test_data.index[sequence_length:], y_pred, label='Predicted', color='orange')

plt.legend()
plt.show()
```
